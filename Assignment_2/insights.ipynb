```python
import time
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tree.base import *
from tree.randomForest import *
from ensemble.ADABoost import *
from ensemble.bagging import *
from ensemble.gradientBoosted import *
X=np.arange(1,1000,50)
#REAL INPUT REAL OUTPUT
np.random.seed(42)
def retTrainTime(N,P):
    X = pd.DataFrame(np.random.randn(N, P))
    y = pd.Series(np.random.randn(N))
    st=time.time()
    dt=AdaBoostClassifier(base_estimator='DecisionTree')
    dt.fit(X,y)
    et=time.time()
    return (et-st)
X=list(X)

for k in [5,50,100]:
    mean=[]
    std=[]
    for i in X:
        y2=[]
        for j in range(5):
            y2.append(retTrainTime(i,k))
        sery2=pd.Series(y2)
        mean.append(sery2.mean())
        std.append(sery2.std())
    plt.plot(X,mean,label=f'AdaBoostClassifier M={k}')
plt.legend()
plt.title('Timing Analysis Of ADABoostClassifier')
plt.xlabel("Number of rows")
plt.ylabel("Time")
```
